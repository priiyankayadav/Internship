# Customer Retention Case Study

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('customer_retention_dataset.xlsx')
```

```python
df
```

# Rows & Columns 

```python
df.columns = df.columns.str.rstrip()
```

```python
print("The data frame has {} rows and {} columns".format(df.shape[0],df.shape[1]))
```

# Data type of each column

```python
#display the data type of each column
for i in df.columns:
    print("The data type of", i,"is",df[i].dtypes)
```

```python
#display the information of data frame
df.info()
```

# Describing data set

```python
#display the 5 point summary
df.describe()
```

# Null values

```python
#check null values are there in each column or not
df.apply(lambda x : sum(x.isnull()))
```

# Unique values

```python
#check how many unique values are there in each column
df.apply(lambda x: len(x.unique()))
```

```python
sns.countplot(x='1Gender of respondent',data=df)
```

From above we can see that females are doing more shopping than males.

```python
ax = sns.countplot(x='2 How old are you?',data=df)

ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.tight_layout()
plt.show()
```

```python
ax = sns.countplot(x = '3 Which city do you shop online from?',data=df)

ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.tight_layout()
plt.show()
```

Here we can deduce that city delhi has the most online customers and city bulandshahr has the least online customers.

```python
ax = sns.countplot(x = '5 Since How Long You are Shopping Online ?', data =df)

ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.tight_layout()
plt.show()
```

Here we can see that above 4 years of age has the most count for online shopping and age between 1-2 years has the lowest count.

```python
df.groupby(['1Gender of respondent','2 How old are you?']).count()
```

```python
ax = sns.countplot(x = '8 Which device do you use to access the online shopping?', data =df)

ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.tight_layout()
plt.show()
```

Smartphone is the most used device to access online shopping and tablet is the least used.

```python
ax = sns.countplot(x = '9 What is the screen size of your mobile device?', data =df)

ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.tight_layout()
plt.show()
```

Other category of mobile device is largely used for online shoppping whereas 5 inches is the least used category of mobile device.

```python
ax = sns.countplot(x = '11 What browser do you run on your device to access the website?', data =df)

ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.tight_layout()
plt.show()
```

Here you can find that goggle chrome browser is mostly run to access the website and mozilla firefox is the least run browser.

```python
ax = sns.countplot(x = '12 Which channel did you follow to arrive at your favorite online store for the first time?', data =df)

ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.tight_layout()
plt.show()
```

Search engine is the channel followed to arrive at the online store for the first time.

```python
ax = sns.countplot(x = '13 After first visit, how do you reach the online retail store?', data =df)

ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.tight_layout()
plt.show()
```

Here we can see that search engine is the largely used path to reach the online retail store.
